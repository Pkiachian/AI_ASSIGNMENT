# International Football Match Analysis

```python
import pandas as pd

df = pd.read_csv("results.csv")
df.head()
```

## Basic Data Exploration

```python
import pandas as pd

# 1. Total number of records
total_matches = len(df)

# 2. Date conversion and range identification
df['date'] = pd.to_datetime(df['date'])
earliest_year = df['date'].dt.year.min()
latest_year = df['date'].dt.year.max()

# 3. Count unique countries
unique_countries = df['country'].nunique()

# 4. Most frequent home team
most_frequent_home_team = df['home_team'].value_counts().idxmax()
home_team_count = df['home_team'].value_counts().max()

# 5. Summary print
print(f'Total matches: {total_matches}')
print(f'Historical range: {earliest_year} to {latest_year}')
print(f'Number of unique countries: {unique_countries}')
print(f'Team with most home matches: {most_frequent_home_team} ({home_team_count} matches)')
```

## Goals Analysis

```python
import pandas as pd

# 1. Create total_goals column
df['total_goals'] = df['home_score'] + df['away_score']

# 2. Average goals per match
avg_total_goals = df['total_goals'].mean()

# 3. Highest-scoring matches
max_goals = df['total_goals'].max()
highest_scoring_matches = df[df['total_goals'] == max_goals]

# 4. Compare total home vs away scores
total_home_goals = df['home_score'].sum()
total_away_goals = df['away_score'].sum()

# 5. Most frequent total goals (mode)
most_frequent_total = df['total_goals'].mode()[0]

print(f'Average goals per match: {avg_total_goals:.2f}')
print(f'Most common number of goals: {most_frequent_total}')
print(f'Total Home Goals: {total_home_goals} | Total Away Goals: {total_away_goals}')
print(f'\nHighest-scoring match(es) ({max_goals} goals):')
print(highest_scoring_matches[['date', 'home_team', 'away_team', 'home_score', 'away_score', 'tournament']])
```

## Match Outcome Categorization

```python
def get_result(row):
    if row['home_score'] > row['away_score']:
        return 'Home Win'
    elif row['away_score'] > row['home_score']:
        return 'Away Win'
    else:
        return 'Draw'

# Apply the function to create the 'result' column
df['result'] = df.apply(get_result, axis=1)

# Calculate percentages
outcome_counts = df['result'].value_counts()
outcome_percentages = (outcome_counts / len(df)) * 100

print("Match Outcome Distribution:")
for result, percentage in outcome_percentages.items():
    print(f"{result}: {percentage:.2f}%")

print(f"\nTotal Matches Analyzed: {len(df)}")
```

## Historical Win Analysis

```python
import pandas as pd

# 1. Identify home wins and extract home_team
home_wins = df[df['result'] == 'Home Win']['home_team']

# 2. Identify away wins and extract away_team
away_wins = df[df['result'] == 'Away Win']['away_team']

# 3. Combine winning teams into a single Series
all_wins = pd.concat([home_wins, away_wins])

# 4. Calculate total number of wins for every country
total_wins = all_wins.value_counts()

# 5. Print the top country and total win count
top_country = total_wins.idxmax()
max_wins = total_wins.max()

print(f'Top winning country: {top_country} with {max_wins} total wins.')
print('\nTop 10 Countries by Total Wins:')
print(total_wins.head(10))
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_theme(style="whitegrid")

# Create a figure with three subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# 1. Histogram for the distribution of total goals
sns.histplot(df['total_goals'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Total Goals per Match', fontsize=14)
axes[0].set_xlabel('Total Goals')
axes[0].set_ylabel('Frequency')

# 2. Bar chart for match outcomes
sns.barplot(x=outcome_counts.index, y=outcome_counts.values, ax=axes[1], palette='viridis', hue=outcome_counts.index, legend=False)
axes[1].set_title('Frequency of Match Outcomes', fontsize=14)
axes[1].set_xlabel('Result')
axes[1].set_ylabel('Number of Matches')

# 3. Bar chart for top 10 teams by total wins
top_10_wins = total_wins.head(10)
sns.barplot(x=top_10_wins.values, y=top_10_wins.index, ax=axes[2], palette='magma', hue=top_10_wins.index, legend=False)
axes[2].set_title('Top 10 Countries by Total Wins Historically', fontsize=14)
axes[2].set_xlabel('Number of Wins')
axes[2].set_ylabel('Country')

plt.tight_layout()
plt.show()
```
